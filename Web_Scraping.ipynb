```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# web scraping from wikipedia

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
# <table class="wikitable sortable jquery-tablesorter" style="text-align:right;">
# <thead><tr>


# same as below
soup.find_all('table', class_ = "wikitable sortable"  )
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
column_name = table.find_all('th')
```

```python
print(column_name)
```

```python
word_title = [title.text.strip() for title in column_name]
print(word_title)
```

```python
df = pd.DataFrame(columns = word_title)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    every_data = [data.text.strip() for data in row_data]
    #print(every_data)
    length =len(df)
    df.loc[length] = every_data
```

```python
df
```

```python
# to export file
df.to_csv(r'C:\Users\Codenester\Desktop\work\web_scrap_file.csv', index = False)
```

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_ratings = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_ratings
```

```python
avg_ratings = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_ratings
```

```python
new_rating = num_ratings.merge(avg_ratings,on='Book-Title')
new_rating
```

```python
popularity = new_rating[new_rating['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popularity = popularity.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_rating','avg_rating','Publisher','Year-Of-Publication']]
```

```python
popularity
```

## colaborative filtering

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
rate_by_user = x[x].index
```

```python
user_rated= ratings_with_name[ratings_with_name['User-ID'].isin(rate_by_user)]
```

```python
y = user_rated.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
filtered_book = user_rated[user_rated['Book-Title'].isin(famous_books)]
```

```python
pt = filtered_book.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
# to find index position
print(np.where(pt.index=='1984')[0][0])
print(np.where(pt.index=='Zoya')[0][0])
```

```python
def recomend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])),key=lambda x: x[1],reverse=True)[1:10]
    
    for i in similar_item:
        print(pt.index[i[0]])
    
```

```python
recomend('The Notebook')
```
